## โปรแกรมหลัก

```python
import random

def read_words():
    d={}
    with open('words.tsv', encoding='utf8') as f:
        for r in f:
    #         print(r, end='')
            lst=r.strip().split('\t')
    #         print(lst)
            d[lst[0]]=lst[1:]
#         print(d)
    return d


def name_meaning(name: str, words):
#     w=[]
#     for c in name.upper():
#     #     print(random.choice(words[c]))
#         w.append(random.choice(words[c]).capitalize())
# #     print('-'.join(w))
#     return '-'.join(w)
    return '-'.join([random.choice(words[c]).capitalize() for c in name.upper()])

# main
words=read_words()
```

```python
# name_meaning('tony', words)

name=input('enter your name: ')
name_meaning(name, words)
```
