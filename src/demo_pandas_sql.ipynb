```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib
matplotlib.style.use('ggplot')
import pyodbc
```

```python
con_string = "driver=ODBC Driver 13 for SQL Server;server=saturn;database=demo;trusted_connection=yes"
```

```python
sql_cmd = """
    select top 10 * 
        from country
        where population > ? and continent = ?
        order by population desc
"""
df = pd.read_sql(sql_cmd, pyodbc.connect(con_string), params=[2000000, 'Africa'])
df.head()
```

```python
df[['Name', 'Population']]
```

```python
df['Population'].plot.bar()
```
