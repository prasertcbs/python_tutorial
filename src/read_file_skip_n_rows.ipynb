```python
skip_n_rows=3
date=[]
value=[]
# with open (r'C:\Users\prasert\Downloads\multiTimeline.csv', encoding='utf8') as f:
with open (r'multiTimeline.csv', encoding='utf8') as f:
    for i, r in enumerate(f):
        if i > skip_n_rows - 1:
#             print(r)
            d, v=r.strip().split(',')
            date.append(d)
            value.append(int(v))
    print(date)
    print('-------------')
    print(value)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(16, 5))
plt.plot(date, value)
```
